# Creating dataframe with latitude and longitude values of PostalCodes

### Following same steps to generate the dataframe with ['PostalCode','Borough','Neighborhood'] as columns

### Then using the geospatial_data.csv file to make a new dataframe and then combining it with the previous dataframe on PostalCode column

```python
import bs4
import requests
import lxml.html as lh
import numpy as np
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
html_data = requests.get(url).text
```

```python
html_data[:100]
```

```python
soup = bs4.BeautifulSoup(html_data,'html.parser')
```

```python
data_table = soup.find('table',{'class':'wikitable sortable'})
```

```python
col=[]
```

```python
for row in data_table.findAll('tr'):
    each_row = row.findAll('td')
    if(len(each_row)==3):
        col.append((each_row[0].text.strip(), each_row[1].text.strip(), each_row[2].text.strip()))                
    
```

```python
col
```

```python
table_array = np.asarray(col)
```

```python
df = pd.DataFrame(table_array)
```

```python
df.head()
```

```python
column_names = ['PostalCode','Borough','Neighborhood']
```

```python
df.columns = column_names
```

```python
df.head()
```

```python
df.shape
```

```python
df = df[df.Borough != 'Not assigned']
```

```python
df.head()
```

```python
df.shape
```

```python
df = df['Neighborhood'].groupby([df.PostalCode, df.Borough]).apply(list).reset_index()
```

```python
df.head()
```

```python
df.shape
```

```python
for x in range(103):
    if(df['Neighborhood'][x][0] == 'Not assigned'):
        df['Neighborhood'][x][0]=df['Borough'][x]    
```

```python
df
```

```python
df.shape
```

## Now we need to read the geospatial_data.csv file and then merge it with the dataframe created above.

Reading the csv file into df_2 

```python
df_2 = pd.read_csv("cor.csv")
```

```python
df_2.head()
```

Assigning new Column names to df_2

```python
df_2.columns = ['PostalCode','Latitude','Longitude']
```

###### merging df and df_2 into final_df based on PostalCode column 

```python
final_df = pd.merge(df,df_2,on='PostalCode')
```
